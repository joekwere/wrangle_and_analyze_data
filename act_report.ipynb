```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df_master = pd.read_csv('twitter_archive_master.csv')
```

```python
df_master.info()
```

## I came up with 4 insights and was able to visualize all 4 of them.

1.	What is the most common breed of dog

As someone that does not know so much about dogs I was really interested in knowing the most common breed of dog that is accepted generally all over the world. Knowing that the data is gathered from the web which is assessable to everyone with internet access.

```python
breed_count = df_master[df_master.breed != 'none'].breed.value_counts().head()
breed_count = breed_count.sort_values(ascending=False)
breed_count
breed_count.plot(figsize = (5,5),kind='bar');
plt.xticks(rotation=90);
plt.title('Top 5 dog breeds tweet count', fontsize=20, fontweight='bold')
plt.ylabel('Number of tweets', fontsize=20);
```

2.	From which source did we get the highest tweet

How were people able to send in their tweets was another subject of interest. The analysis showed that most people used iphone and that accounted for over 98% of the tweets that were analyzed. 

```python
fig = plt.figure(figsize =(7, 7))
plt.pie(df_master['source'].value_counts().values, autopct='%.1f%%', labels = ['iphone', 'web', 'deck']);
```

3.	Top 5 breeds of dogs with the highest likes(favorite_count)

I was also interested in knowing the breed of dog that is liked most and also if that is the most common brand of dog. From the analysis, the dog that is most liked is the golden retriever which is also the most common breed of dog. This analysis was done excluding the dogs that their breed is unknown

```python
fav_counts_breed = df_master[df_master.breed != 'none']
fav_counts_breed = fav_counts_breed.groupby(['breed'])
fav_counts_breed = fav_counts_breed['favorite_count'].sum()
fav_counts_breed = fav_counts_breed.sort_values(ascending = False)
top5_breeds_fav_counts = fav_counts_breed.head()
top5_breeds_fav_counts
top5_breeds_fav_counts.plot(figsize = (5,5),kind='bar');
plt.xticks(rotation=90);
plt.title('Favorite breed of dog', fontsize=20, fontweight='bold')
```

4.	Top 5 preffered dog names

Just in a case I decide to get a dog tomorrow, let me find out what names people prefer to call their dogs. It was interesting to note that Charlie, Zoey and Lucy were among the first 5 preferred names for dogs. Well, I have friends bearing these 3 names so I am not getting a dog.

```python
name_count = df_master[df_master.name != 'None']
name_count = name_count.groupby(['name'])['favorite_count'].sum()
name_count = name_count.sort_values(ascending = False)
top5_name_count = name_count.head()
top5_name_count
top5_name_count.plot(figsize = (5,5),kind='bar');
plt.xticks(rotation=90);
plt.title('Most preffered dog name', fontsize=20, fontweight='bold')
```
